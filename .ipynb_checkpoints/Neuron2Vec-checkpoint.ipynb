```python
def get_cluster_distance(centers):
    cluster_distance = np.empty((centers.shape))
    for i in range(len(centers)):
        cluster_distance[i,:] = np.array([np.abs(np.linalg.norm(centers[i] - center)) for center in centers])
    return cluster_distance

def get_cluster_pair(cluster_distance, pair_num, far=True):
    cluster_pair = np.empty((cluster_distance.shape[0], pair_num))
    for i in range(len(cluster_distance)):
        if far:
            cluster_pair[i,:] = np.argsort(cluster_distance[i])[::-1][:pair_num]
        else:
            cluster_pair[i,:] = np.argsort(cluster_distance[i])[1:pair_num+1]
    return cluster_pair
```

```python
def make_img_html(imgsrc_list, imgsrc_list_far, imgsrc_list_close):
    close_head = '<div class="col" id="column1"><h1>Similar Group</h1>'
    far_head = '<div class="col" id="column2"><h1>Dissimilar Group</h1>'
    target_head = '<div class="col" id="column3"><h1>Target Group</h1>'
    target_note = '<h1>Write properties of this Group</h1>'
    div_str = '</div>'
    html_str = close_head 
    for imgsrc in imgsrc_list_close:
        html_str += make_img_div(imgsrc)
    html_str += div_str+target_head
    for imgsrc in imgsrc_list:
        html_str += make_img_div(imgsrc)
    html_str += target_note+div_str +far_head 
    for imgsrc in imgsrc_list_far:
        html_str += make_img_div(imgsrc)
    html_str += div_str
    return html_str
    
def make_img_div(imgsrc):
    return '<img src="../.'+imgsrc+'">'
```
